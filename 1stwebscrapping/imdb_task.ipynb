## Web scrapping de IMDB

Voy a descargar los datos del [top 250 películas de imbd](https://www.imdb.com/chart/top) a través de webscraping. Crearé un dataframe con esa información y posteriormente lo guardaré en un csv.

Doy la siguiente información:
* Título película
* Año estreno película
* Posición ranking
* Reparto principal
* Rating estrellas
* Nº de votos

```python
from bs4 import BeautifulSoup as bs
import requests
import pandas as pd
# from splinter import Browser
import numpy as np
```

```python
url = "https://www.imdb.com/chart/top"
response = requests.get(url)
```

```python
html = response.content
```

```python
soup = bs(html, 'html.parser')
```

```python
print(soup)
```

```python
var_p = soup.find("td", class_="titleColumn")
```

```python
print(var_p)
```

```python
all_position = soup.find_all("td", class_="titleColumn") #posicion en el ranking
for td in all_position:
    print(td.get_text(strip=True).split('.')[0]) #el split cuando separa hacer dos lados
```

```python
all_p = soup.find_all("td", class_="titleColumn") # solo el nombre de las pelis
for td in all_p:
    print(td.find('a').get_text())
```

```python
all_rat = soup.find_all("td", class_="ratingColumn imdbRating") #ratings pero con espacios
for td in all_rat:
    print(td.find('strong').get_text())
```

```python
all_prot = soup.find_all("td", class_="titleColumn") #reparto
for td in all_prot:
    print(td.find('a').get('title'))
```

```python
all_vot = soup.find_all("td", class_="ratingColumn imdbRating") #votos
for td in all_vot:
    print(td.find('strong').get('title').split(' ')[3]) 
    #dividir toda la frase que había por espacios y coger la posición en la que estaban los votos
```

```python
all_year = soup.find_all("span", class_="secondaryInfo") #año
for x in all_year:
    print(x.get_text())
```

```python
top_250_dict = {
    "Título película": [],
    "Año estreno película": [],
    "Posición ranking": [],
    "Reparto principal": [],
    "Rating estrellas": [],
    "Nº de votos": []
}

all_position = soup.find_all("td", class_="titleColumn")
for td in all_position:
    top_250_dict["Posición ranking"].append(td.get_text(strip=True).split('.')[0])

all_p = soup.find_all("td", class_="titleColumn")
for td in all_p:
    top_250_dict["Título película"].append(td.find('a').get_text())

all_rat = soup.find_all("td", class_="ratingColumn imdbRating")
for td in all_rat:
    top_250_dict["Rating estrellas"].append(td.find('strong').get_text())

all_prot = soup.find_all("td", class_="titleColumn")
for td in all_prot:
    top_250_dict["Reparto principal"].append(td.find('a').get('title'))

all_vot = soup.find_all("td", class_="ratingColumn imdbRating")
for td in all_vot:
    top_250_dict["Nº de votos"].append(td.find('strong').get('title').split(' ')[3])

all_year = soup.find_all("span", class_="secondaryInfo")
for x in all_year:
    top_250_dict["Año estreno película"].append(x.get_text())

df = pd.DataFrame(top_250_dict)
```

```python
df.to_csv('250imdb') #para crear un archivo csv del df
```
